##데이터 병합



```python
transaction_2= pd.read_csv('/content/gdrive/MyDrive/데이터분석 스터디/1장/transaction_2.csv')

transaction = pd.concat([transaction_1, transaction_2],axis=0,ignore_index = True)
transaction
# ignore_index=True -> 이전 데이터 인덱스 무시하고 새로 부여

#데이터 올바르게 병합되었는지 확인
# print(len(transaction_1))
# print(len(transaction_2))
# print(len(transaction))
```

```python
transaction_detail_2= pd.read_csv('/content/gdrive/MyDrive/데이터분석 스터디/1장/transaction_detail_2.csv')

transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2],axis=0, ignore_index=True)
transaction_detail.tail(20)
```

```python
join_data = pd.merge(transaction_detail, transaction[['transaction_id','payment_date','customer_id']]
                     ,on = 'transaction_id',how='left')
join_data

# 위에서 transaction_detail 과 transaction 은 데이터 개수가 다르다! 
# 결측치는 없었으니 즉 데이터 개수가 더 많은 transaction_detail에 'transaction_id' 중복값이 있다는 것! (위에서 확인 가능)
# 같은 'transaction_id'에 다른 'item_id'를 가진 데이터가 있기 때문에 데이터 개수 차이 발생
# -> 이것을 merge 를 통해 결합함으로써 엄연히 다른 두 데이터에 모두 transaction 의 정보 붙여주는 것임

# transaction_id를 기준으로 가로 병합. transaction_detail에다가 병합시키는것이기 때문에 left join
```

```python
# 데이터 잘 병합되었는지 확인
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

```python
# 만들어놓은 join_data 에 customer_master 과 item_master 에 포함된 모든 정보 추가해서 새로 제작

join_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head()
```

```python
#매출 열 다시 추가

join_data['price'] = join_data['quantity'] * join_data['item_price']
join_data[['quantity','item_price','price']]
```

```python
#payment_date를 datetime 형태로 변환하여 작업

join_data['payment_date'] = pd.to_datetime(join_data['payment_date'])
join_data['payment_month'] = join_data['payment_date'].dt.strftime('%Y%m')
join_data[['payment_date','payment_month']].head()
```

```python
#월별 매출

join_data.groupby('payment_month').sum()['price']
```

```python
# 월별, 제품별 매출의 합, 수량

join_data.groupby(['payment_month','item_name']).sum()[['price','quantity']]
```

```python
#pivot 형태로 재배치
# aggfunc는 집계방법을 지정
pd.pivot_table(join_data, index = 'item_name', columns='payment_month'
,values = ['price','quantity'],aggfunc='sum')
```

##시각화

```python
graph_data = pd.pivot_table(join_data, index = 'payment_month', columns='item_name',values='price',aggfunc='sum')
graph_data.head()
```

```python
# plt.plot(가로축,세로축)

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(list(graph_data.index),graph_data['PC-A'],label='PC-A')
plt.plot(list(graph_data.index),graph_data['PC-B'],label='PC-B')
plt.plot(list(graph_data.index),graph_data['PC-C'],label='PC-C')
plt.plot(list(graph_data.index),graph_data['PC-D'],label='PC-D')
plt.plot(list(graph_data.index),graph_data['PC-E'],label='PC-E')
plt.legend()
```
